```python

unique_items = purchase_data["Item ID"].nunique()
average = purchase_data["Price"].mean()
number_of_purchases = purchase_data.count()
total_revenue = purchase_data["Price"].sum()

purchasing_analysis = pd.DataFrame({
    'Number of Unique Items': unique_items,
    'Average Price': average,
    'Number of Purchases': number_of_purchases,
    'Total Revenue': total_revenue
}) 
purchasing_analysis.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 10,14,19,24,29,34,39,100000]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34",
                "35-39", "40+"]

purchase_data["Age Buckets"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)
age_group_df = purchase_data.groupby("Age Buckets")

# Find how many rows fall into each bin
#print(age_group_df["SN"].count())

# Get the average of each column within the GroupBy object
#age_group_df[["comments", "duration", "languages"]].mean()

age_counts = age_group_df['Age Buckets'].value_counts()
age_percentages = age_group_df['Age Buckets'].value_counts()/purchase_data['Gender'].count()*100


age_table = pd.DataFrame({
    'Count': age_counts,
    'Percent': age_percentages,
}) 

age_table.head()

```
